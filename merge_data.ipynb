```python
eeg_data = pd.DataFrame()
```

```python
channels = ['ch_'+str(i) for i in range(1,15)]
```

```python
# import required module
import os
 
# assign directory
directory = '25-users'
l = -1
# iterate over files in
# that directory
for filename in os.scandir(directory):
    l+=1
    data = pd.read_csv(filename.path, sep=" ", header=None, names=channels)
    data['label'] = labels[l]
    eeg_data = eeg_data.append(data , ignore_index=True)
```
