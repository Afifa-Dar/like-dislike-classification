```python
eeg_data = pd.DataFrame()
```

```python
channels = ['ch_'+str(i) for i in range(1,15)]
```

```python
# import required module
import os
 
# assign directory
directory = '25-users'
l = -1
# iterate over files in
# that directory
for filename in os.scandir(directory):
    l+=1
    data = pd.read_csv(filename.path, sep=" ", header=None, names=channels)
    data['label'] = labels[l]
    eeg_data = eeg_data.append(data , ignore_index=True)
```

```python
data_without_label = eeg_data.drop('label',axis=1)
```

```python
#applying bandpass filter
import scipy.signal as signal

# Define the filter parameters
fs = 128  # Sampling frequency
lowcut = 0.5  # Lower cutoff frequency
highcut = 30  # Higher cutoff frequency
order = 4  # Filter order

# Calculate the Nyquist frequency
nyq = 0.5 * fs

# Calculate the filter cutoff frequencies in normalized units
low = lowcut / nyq
high = highcut / nyq

# Design a Butterworth bandpass filter
b, a = signal.butter(order, [low, high], btype='band')

# Apply the filter to the EEG data using lfilter
filtered_data = signal.lfilter(b, a, data_without_label)
```

```python
#compare orignal data and filtered data
plt.figure(figsize=(20,10))
plt.plot(data_without_label , color = 'red')
plt.plot(filtered_data , color = 'green')
```

```python
plt.figure(figsize=(20,10))
plt.plot(filtered_data)  #plot filtered data
```

```python
#cpnvert filtered data in dataframe
clean_data = pd.DataFrame(filtered_data, columns=channels)
# clean_data
```

```python
plt.figure(figsize=(30,30))
for i , col in enumerate(channels,1):
    plt.subplot(4,4,i)
    plt.plot(clean_data[col])  #channel wise plot of filtered data
```
