```python
data_1= pd.read_csv('25-users/Abhishek_1.txt', sep=" ", header=None, names=channels)
data_2 = pd.read_csv('25-users/Abhishek_2.txt', sep=" ", header=None, names=channels)
data_3 = data = pd.read_csv('25-users/Abhishek_10.txt', sep=" ", header=None, names=channels)
data_4 = pd.read_csv('25-users/Abhishek_11.txt', sep=" ", header=None, names=channels)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.plot( t , data_1 ) #Abhishek_1.txt
plt.title('like_1')
plt.subplot(2,2,2)
plt.plot( t , data_2 ) #Abhishek_2.txt
plt.title('like_2')
plt.subplot(2,2,3)
plt.plot( t ,  data_3 ) #Abhishek_10.txt
plt.title('dislike_1')
plt.subplot(2,2,4)
plt.plot( t , data_4 ) #Abhishek_11.txt
plt.title('dislike_2')
```

```python
#all channels data of 'data' ie first data file (Abhishek_1.txt)
plt.figure(figsize=(30,30))
for i , col in enumerate(channels,1):
    plt.subplot(4,4,i)
    plt.plot(t , data_1[col])
```
